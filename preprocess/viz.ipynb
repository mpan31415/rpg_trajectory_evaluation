```python
import numpy as np
import os
import matplotlib.pyplot as plt

gt_dir = os.getcwd() + "\\ground_truth_files"
```

### 1. KITTI

```python
kitti_gt_file = gt_dir + '/kitti.txt'
kitti_gt_traj = np.loadtxt(kitti_gt_file, delimiter=' ')

kitti_x = kitti_gt_traj[:, 3]
kitti_z = kitti_gt_traj[:, -1]

# plot
plt.plot(kitti_x, kitti_z, label='kitti')
```

### 2. MALAGA

```python
malaga_gt_file = gt_dir + '/malaga.txt'
malaga_gt_traj = np.genfromtxt(malaga_gt_file, delimiter=None, skip_header=1)

malaga_x = malaga_gt_traj[:, 8]
malaga_z = malaga_gt_traj[:, 9]

# plot
plt.plot(malaga_x, malaga_z, label='malaga')
```
