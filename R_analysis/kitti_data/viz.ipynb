```python
from pandas import read_csv
import matplotlib.pyplot as plt

%matplotlib inline
```

### Load Data

```python
data_df = read_csv("all_percentages.csv")

data_df.head()
```

### Percentage of Tracked Keypoints

```python
detectors = ["sift", "harris", "shi_tomasi"]

for det in detectors:
    det_group = data_df[data_df["detector"]==det]
    tracked_lst = det_group["tracked"].values
    plt.plot(tracked_lst, label=det)
plt.legend()
```

### Percentage of Inliers

```python
detectors = ["sift", "harris", "shi_tomasi"]

for det in detectors:
    det_group = data_df[data_df["detector"]==det]
    inliers_lst = det_group["inliers"].values
    plt.plot(inliers_lst, label=det)
plt.legend()
```

### Detection Time

```python
detectors = ["sift", "harris", "shi_tomasi"]
time_stats_lst = []

for det in detectors:
    det_group = data_df[data_df["detector"]==det]
    time_lst = det_group["detect_time"].values
    mean, std = time_lst.mean(), time_lst.std()
    time_stats_lst.append((mean, std))
    print("Detector: %s, mean: %f, std: %f"%(det, mean, std))
```
